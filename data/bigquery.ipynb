```python
%%bigquery
SELECT
	* EXCEPT (job_highlights_qualifications, job_highlights_responsibilities, job_highlights_benefits)
FROM
	`job-listings-366015`.gsearch_job_listings_clean.gsearch_jobs_fact
LIMIT 10
```

```python
%%bigquery
-- Job posting websites
SELECT
	job_via AS job_website,
	COUNT(job_via) AS job_count,
FROM
	`job-listings-366015`.gsearch_job_listings_clean.gsearch_jobs_fact
GROUP BY
	job_via
ORDER BY
	job_count DESC
```

```python
%%bigquery 
-- Top job titles
SELECT
	job_title,
	COUNT(job_title) AS job_count,
FROM
	`job-listings-366015`.gsearch_job_listings_clean.gsearch_jobs_fact
GROUP BY
	job_title
ORDER BY
	job_count DESC
```

```python
%%bigquery 
-- Degree requirements
SELECT
	search_term,
	ROUND(COUNT(job_no_degree_mention) * 100 / COUNT(search_term), 1) 
       AS no_degree_percent,
	COUNT(job_no_degree_mention) AS no_degree_count,
	COUNT(search_term) AS search_count
FROM
	`job-listings-366015`.gsearch_job_listings_clean.gsearch_jobs_fact
GROUP BY
	search_term
ORDER BY
	no_degree_percent DESC
```

```python
%%bigquery 
-- Work from home
SELECT
	search_term,
	ROUND(COUNT(job_work_from_home) * 100 / COUNT(search_term),
	1) AS wfh_percent,
	COUNT(job_work_from_home) AS wfh_count,
	COUNT(search_term) AS search_count
FROM
	`job-listings-366015`.gsearch_job_listings_clean.gsearch_jobs_fact
GROUP BY
	search_term
ORDER BY
	wfh_percent DESC
```

```python
%%bigquery 

-- Job location
SELECT
	job_location,
	COUNT(job_location) AS job_location_count,
FROM
	`job-listings-366015`.gsearch_job_listings_clean.gsearch_jobs_fact
GROUP BY
	job_location
ORDER BY
	job_location_count DESC
```

```python
%%bigquery 

-- Job type
SELECT
	a1.job_schedule_type,
	ROUND(COUNT(a1.job_schedule_type) * 100 /
       (
	SELECT
		COUNT(job_schedule_type)
	FROM
		`job-listings-366015`.gsearch_job_listings_clean.gsearch_jobs_fact), 1) AS job_type_percent,
	COUNT(job_schedule_type) AS job_type_count
FROM
	`job-listings-366015`.gsearch_job_listings_clean.gsearch_jobs_fact AS a1
GROUP BY
	job_schedule_type
ORDER BY
	job_type_percent DESC
```

```python
%%bigquery 
-- Skills
SELECT
	job_description,
	job_salary,
FROM
	`job-listings-366015`.gsearch_job_listings_clean.gsearch_jobs_fact
WHERE
	job_salary IS NOT NULL
LIMIT 10
```

```python
%%bigquery 
-- salary table
SELECT
	*
FROM
	`job-listings-366015`.gsearch_job_listings_clean.gsearch_salary
LIMIT 10

-- average salary
SELECT
	ROUND(AVG(salary_year)) AS avg_annual_salary,
	ROUND(AVG(salary_hour), 2) AS avg_hourly_salary,
FROM
	`job-listings-366015`.gsearch_job_listings_clean.gsearch_salary
```

```python
%%bigquery 

-- average salary per job title
SELECT
	search_term,
	ROUND(AVG(salary_year)) AS avg_salary_annual,
	-- AVG(salary_hour) AS avg_salary_hourly
FROM
	`job-listings-366015`.gsearch_job_listings_clean.gsearch_salary AS d
LEFT JOIN `job-listings-366015`.gsearch_job_listings_clean.gsearch_jobs_fact AS f ON
	f.job_id = d.job_id
GROUP BY
	search_term
ORDER BY
	avg_salary_annual DESC
```

```python
%%bigquery 

-- range salary
SELECT
	search_term,
	MAX(salary_max - salary_min) AS salary_range,
FROM
	`job-listings-366015`.gsearch_job_listings_clean.gsearch_salary AS d
LEFT JOIN `job-listings-366015`.gsearch_job_listings_clean.gsearch_jobs_fact AS f ON
	f.job_id = d.job_id
GROUP BY
	search_term
ORDER BY
	salary_range DESC
```

```python
%%bigquery 

-- top salary range inspect
SELECT
	* EXCEPT (job_highlights_qualifications, job_highlights_responsibilities, job_highlights_benefits)
FROM
	`job-listings-366015`.gsearch_job_listings_clean.gsearch_salary AS d
LEFT JOIN `job-listings-366015`.gsearch_job_listings_clean.gsearch_jobs_fact AS f ON
	f.job_id = d.job_id
WHERE
	(salary_max - salary_min) > 250000
```

```python
%%bigquery salary --project job-listings-366015

-- median salary
SELECT
	search_term,
	median_salary_annual,
	median_salary_hourly,
FROM
	(
	SELECT
		search_term,
		PERCENTILE_CONT(salary_year, 0.5) OVER (PARTITION BY search_term)
                 AS median_salary_annual,
		PERCENTILE_CONT(salary_hour, 0.5) OVER (PARTITION BY search_term)
                 AS median_salary_hourly
	FROM
		`job-listings-366015`.gsearch_job_listings_clean.gsearch_salary AS d
	LEFT JOIN `job-listings-366015`.gsearch_job_listings_clean.gsearch_jobs_fact AS f ON
		f.job_id = d.job_id)
GROUP BY
	search_term,
	median_salary_annual,
	median_salary_hourly
```
